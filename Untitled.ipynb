```python
import torch
import matplotlib.pyplot as plt
```

```python
from dataloader import *
from torch.utils.data import DataLoader
```

```python
source = 'modelnet_11'
target = 'scanobjectnn_11'
swapax = source == 'modelnet_11'
batch_size=16
```

```python
def visualize_pc(data, idx=0):
    data = data.detach().cpu().numpy()
    xdata = data[idx][0, :].squeeze()
    ydata = data[idx][1, :].squeeze()
    zdata = data[idx][2, :].squeeze()
    
    fig = plt.figure(figsize=(15, 15))
    ax = plt.axes(projection='3d')
    
    ax.scatter3D(xdata, ydata, zdata, marker='o')
    plt.show()
```

```python
source_train_dataset = PaddingData(pc_input_num=2048, status='train', aug=True, swapax=swapax, pc_root='/ws/data/%s' % source)
source_train_dataloader = DataLoader(source_train_dataset, batch_size=batch_size, shuffle=True, num_workers=4, drop_last=True)
```

```python
data, pn, label = next(iter(source_train_dataloader))
```

```python
visualize_pc(data)
```
